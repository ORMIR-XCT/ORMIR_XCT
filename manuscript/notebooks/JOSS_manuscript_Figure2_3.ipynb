# Comparing IPL and ORMIR_XCT Calculated Thicknesses and BMD

To compare the IPL and ORMIR_XCT algorithms for thickness estimation and bone mineral density (BMD) calculation, Bland-Altman and linear regression plots were generated using data from simulated shapes. The shapes used are freely available on Zenodo (https://doi.org/10.5281/zenodo.8155738) and include:
- Hollow spheres
- Solid speheres
- Hollow cylinders
- Solid cylinders
- Plates

CSV files containing the calculated thickness of each shape using IPL and ORMIR_XCT are provided in the csv direcotry.
